```python
import pandas as pd
```

```python
#reading the file
pf = pd.read_csv(r"C:\Users\Vedant Kadu\OneDrive\Desktop\airbnb_data_cleaning\Airbnb_Open_Data.csv",encoding="utf-8")
print(pf)
```

```python
#checking the describe, head, coloums 
print(pf.head())
print(pf.columns)
print(pf.describe)
```

```python
#checking the nul values
print(pf.isnull().sum())
```

```python
#dropping unecessary coloums
pf.drop(columns=["reviews per month", "review rate number", "calculated host listings count", "availability 365","house_rules","license"], inplace=True)
print(pf)
```

```python
#printing the coloums
print(pf.columns)
```

```python
#modifying the name 
pf.rename(columns={"NAME":"name"}, inplace=True)
print(pf.columns)
```

```python
#lets like if you want to make the coloums name bigger so here is the code
new_coloum = []
for i in pf.columns:
    new_coloum.append(i.upper())

#copying all the content from the pf to df
df = pf.copy()
df.columns = new_coloum
print(df)
```

```python
#checking the duplicate values
print(df.duplicated().sum())
```

```python
#droping the duplicate values
df = df.drop_duplicates()
print(df.duplicated().sum())
```

```python
#removing the nan values
print(df.isna().sum())
```

```python
#dropping the last review coloum

df.drop(columns=["LAST REVIEW"], inplace=True)
print(df.isna().sum())
```

```python
#modifying the name coloum
df["NAME"] = df["NAME"].str.title()
df["NAME"] = df["NAME"].str.replace(r"[^a-zA-Z0-9\s]","", regex=True)
```

```python
#dropping and modifiying the na value
df.dropna(subset=["NAME", "HOST_IDENTITY_VERIFIED", "HOST NAME", "NEIGHBOURHOOD GROUP", "NEIGHBOURHOOD", "LAT", "LONG"], inplace=True)
print(df.columns)

#Modifying the coloums here
df["COUNTRY CODE"] = df["COUNTRY CODE"].fillna("XX")
df["COUNTRY"]= df["COUNTRY"].fillna("Unknown")
df["INSTANT_BOOKABLE"]= df["INSTANT_BOOKABLE"].fillna("f")
df["CANCELLATION_POLICY"] = df["CANCELLATION_POLICY"].fillna("moderate")
df["NUMBER OF REVIEWS"]= df["NUMBER OF REVIEWS"].fillna(0)

#modifying the coloums with the median
df["CONSTRUCTION YEAR"]= df["CONSTRUCTION YEAR"].fillna(df["CONSTRUCTION YEAR"].median())
df["MINIMUM NIGHTS"] = df["MINIMUM NIGHTS"].fillna(df["MINIMUM NIGHTS"].median())
```

```python
#modifying the specific coloums
df["PRICE"] = df["PRICE"].astype(str).replace(r"[\$,]", "", regex=True).astype(float)
df["SERVICE FEE"] = df["SERVICE FEE"].astype(str).replace(r"[\$,]", "", regex=True).astype(float)
```

```python
#filling the na values
df["PRICE"] = df["PRICE"].fillna(df["PRICE"].median())
df["SERVICE FEE"]= df["SERVICE FEE"].fillna(df["SERVICE FEE"].median())
```

```python
#applying the lambda function 
"""
for this we can also also use this line
df["INSTANT_BOOKABLE"].astype(int)
"""
df["INSTANT_BOOKABLE"] = df["INSTANT_BOOKABLE"].apply(lambda x: 1 if x==True else 0)
df["INSTANT_BOOKABLE"]
```
